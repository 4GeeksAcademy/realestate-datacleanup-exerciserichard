#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
# Definir las poblaciones del "cinturón sur" de Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original usando la columna 'level5' y la función 'isin'
south_belt_ds = ds[ds['level5'].isin(south_belt_populations)]

# Normalizar la variable de precios para cada población
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

south_belt_ds['normalized_price'] = south_belt_ds.groupby('level5')['price'].transform(normalize)

# Crear el gráfico de histogramas
plt.figure(figsize=(10, 6))

for population in south_belt_populations:
    subset = south_belt_ds[south_belt_ds['level5'] == population]
    plt.hist(subset['normalized_price'], bins=20, alpha=0.5, label=population)

plt.title('Normalized Price Histograms for South Belt Populations')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO
# Definir las poblaciones del "cinturón sur" de Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original usando la columna 'level5' y la función 'isin'
south_belt_ds = ds[ds['level5'].isin(south_belt_populations)]

# Crear una nueva columna 'pps' (price per square meter)
south_belt_ds['pps'] = south_belt_ds['price'] / south_belt_ds['surface']

# Calcular la media y la varianza de 'pps' en "Getafe"
getafe_mean_pps = south_belt_ds[south_belt_ds['level5'] == 'Getafe']['pps'].mean()
getafe_var_pps = south_belt_ds[south_belt_ds['level5'] == 'Getafe']['pps'].var()

# Calcular la media y la varianza de 'pps' en "Alcorcón"
alcorcon_mean_pps = south_belt_ds[south_belt_ds['level5'] == 'Alcorcón']['pps'].mean()
alcorcon_var_pps = south_belt_ds[south_belt_ds['level5'] == 'Alcorcón']['pps'].var()

# Imprimir los valores obtenidos
print(f'Getafe - Mean PPS: {getafe_mean_pps:.2f} USD/m², Variance PPS: {getafe_var_pps:.2f}')
print(f'Alcorcón - Mean PPS: {alcorcon_mean_pps:.2f} USD/m², Variance PPS: {alcorcon_var_pps:.2f}')
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:

import folium


# Definir las poblaciones del "cinturón sur" de Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original usando la columna 'level5' y la función 'isin'
south_belt_ds = ds[ds['level5'].isin(south_belt_populations)]

# Convertir las coordenadas de cadena a float
south_belt_ds['latitude'] = south_belt_ds['latitude'].str.replace(',', '.').astype(float)
south_belt_ds['longitude'] = south_belt_ds['longitude'].str.replace(',', '.').astype(float)

# Crear un diccionario de colores para cada población
colors = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

# Crear un mapa centrado en el "cinturón sur" de Madrid
m = folium.Map(location=[40.35, -3.75], zoom_start=12)

# Añadir puntos al mapa
for _, row in south_belt_ds.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color=colors[row['level5']],
        fill=True,
        fill_color=colors[row['level5']],
        fill_opacity=0.6,
        popup=row['level5']
    ).add_to(m)

# Mostrar el mapa
m.save('south_belt_map.html')
m
```
